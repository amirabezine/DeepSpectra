```python
import torch
import psutil


def get_memory_usage():
    process = psutil.Process()
    return process.memory_info().rss / (1024*1024)  # in MB

before = get_memory_usage()
# Before optimization
print(f"Memory usage before: {get_memory_usage():.2f} MB")
# Your large PE vector
PE = torch.randn(200000)

# List of smaller Z vectors
Z_list = [torch.randn(20) for _ in range(10000)]  # Example with 5 Z vectors

# Function to combine a Z vector with PE
def combine_Z_with_PE(Z, PE):
    return torch.cat([Z, PE])

# Iterate through Z vectors and combine each with PE
results = []
for Z in Z_list:
    combined = combine_Z_with_PE(Z, PE)
    results.append(combined)
    print(results.shape)

# If you need to stack all results
stacked_results = torch.stack(results)
after = get_memory_usage()
print(f"Memory usage after: {get_memory_usage():.2f} MB")
print(after - before)
```
