```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Path to your HDF5 file
file_path = 'healpixfiles/spectra_healpix_170.hdf5'

# Open the HDF5 file
with h5py.File(file_path, 'r') as f:
    # Access the groups
    group1 = f['170_combined_10']
    group2 = f['170_combined_11']
    
    # Retrieve the wavelength tensors
    wavelengths1 = group1['wavelength'][:]
    wavelengths2 = group2['wavelength'][:]
    
    # Ensure both tensors have the same length
    if len(wavelengths1) != len(wavelengths2):
        raise ValueError("The wavelength tensors have different lengths")
    
    # Identify the indices where the wavelengths differ
    differences = wavelengths1 != wavelengths2
    differing_indices = np.where(differences)[0]
    
    # Plot the differences
    plt.figure(figsize=(12, 6))
    plt.plot(wavelengths1, label='Group 170_combined_10')
    plt.plot(wavelengths2, label='Group 170_combined_11')
    plt.scatter(differing_indices, wavelengths1[differing_indices], color='red', label='Differences')
    plt.title('Wavelength Differences Between Groups 170_combined_10 and 170_combined_11')
    plt.xlabel('Index')
    plt.ylabel('Wavelength')
    plt.legend()
    plt.show()
    
    # Print the number of differing wavelengths
    num_differences = np.sum(differences)
    print(f"Number of differing wavelengths: {num_differences}")
```
