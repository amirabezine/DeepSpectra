```python
import torch
from torch.utils.data import Dataset, DataLoader
from astropy.io import fits
import os
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
import random
```

```python
class GALAHDR3SpectraDataset(Dataset):
    def __init__(self, directory, max_files=None):
        """
        Args:
            directory (string): Directory with all the FITS files.
            max_files (int): Maximum number of FITS files to load (optional).
        """

        all_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith('.fits')]
        if max_files is not None and max_files < len(all_files):
            self.files = random.sample(all_files, max_files)  # Randomly select max_files from the list
        else:
            self.files = all_files

    def __len__(self):
        """
        Returns the number of files in the dataset.
        """
        return len(self.files)
    
    def __getitem__(self, idx):
        """
        Retrieves different spectral data and their wavelength grids from a FITS file at the specified index.

        Parameters:
        - idx: int, index of the FITS file in the list.

        Returns:
        - data_dict: dict, containing the wavelength and flux for each type of spectral data.
        """
        with fits.open(self.files[idx]) as hdul:
            data_dict = {}
            for hdu_index, hdu_name in enumerate(["input_sigma", "no_sky_subspectrum", "no_sky_sigma", "normalised spectra"]):
                flux = hdul[hdu_index + 1].data.astype(np.float32)
                header = hdul[hdu_index + 1].header
                
                # Compute the wavelength grid
                w_start = header['CRVAL1']  # Starting wavelength
                w_delta = header['CDELT1']  # Wavelength increment per pixel
                n_pixels = len(flux)
                wavelength = w_start + w_delta * np.arange(n_pixels)  # Wavelength array

                # Convert to PyTorch tensors
                flux = torch.from_numpy(flux)
                wavelength = torch.from_numpy(wavelength)

                data_dict[hdu_name] = (wavelength, flux)

        return data_dict
```

```python
directory = '../../projects/k-pop/spectra/galah/dr3'
dataset = GALAHDR3SpectraDataset(directory, max_files=100)
dataloader = DataLoader(dataset, batch_size=10, shuffle=True)
```

```python
import matplotlib.pyplot as plt
import random

def plot_spectra(spectra_data):
    """
    Plots multiple spectra from different HDUs.

    Parameters:
    - spectra_data: dict, contains tuples of (wavelength, flux) for different spectra types.
    """
    plt.figure(figsize=(14, 10))
    for i, (key, (wavelength, flux)) in enumerate(spectra_data.items()):
        plt.subplot(2, 2, i + 1)
        plt.plot(wavelength.numpy(), flux.numpy(), label=f'{key}', color='b')
        plt.title(f'Spectrum Type: {key}')
        plt.xlabel('Wavelength (Angstrom)')
        plt.ylabel('Flux')
        plt.legend()
        plt.grid(True)
    
    plt.tight_layout()
    plt.show()

# To use the plotting function:
random_index = np.random.randint(len(dataset))  # Get a random index
spectra_data = dataset[random_index]
plot_spectra(spectra_data)
```

```python

def plot_modified_spectra(dataset, index=None):
    """
    Plots the modified spectra by subtracting and adding the input sigma from/to the normalized spectra
    from a selected FITS file in the dataset.

    Parameters:
    - dataset: instance of GALAHDR3SpectraDataset
    - index: int, optional index to specify a particular FITS file. If None, a random file is chosen.
    """
    if index is None:
        index = random.randint(0, len(dataset) - 1)  # Select a random index if none provided

    # Load the data for the specified index
    spectra_data = dataset[index]

    # Extract wavelength and flux data
    wavelength = spectra_data['normalised spectra'][0].numpy()
    normalised_flux = spectra_data['normalised spectra'][1].numpy()
    input_sigma_flux = spectra_data['input_sigma'][1].numpy()

    # Calculate the modified spectra
    spectrum_minus_sigma = normalised_flux - input_sigma_flux
    spectrum_plus_sigma = normalised_flux + input_sigma_flux

    # Create the plot
    plt.figure(figsize=(10, 5))
    plt.plot(wavelength, spectrum_minus_sigma, label='Normalized - Input Sigma', color='red', linewidth=0.5)
    plt.plot(wavelength, spectrum_plus_sigma, label='Normalized + Input Sigma', color='green', linewidth=0.5)
    plt.title(f'Modified Normalized Spectra for File Index: {index}')
    plt.xlabel('Wavelength (Angstrom)')
    plt.ylabel('Flux')
    plt.legend()
    plt.grid(True)
    plt.show()



plot_modified_spectra(dataset) 
```
