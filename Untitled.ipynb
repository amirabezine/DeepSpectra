```python
import torch
from torch.utils.data import Dataset
from astropy.io import fits
import matplotlib.pyplot as plt

class APOGEEBitMaskDataset(Dataset):
    def __init__(self, filepath, hdu_index=3):
        """
        Initialize the dataset with the path to the FITS file and the index of the HDU containing the bitmask.
        """
        self.filepath = filepath
        self.hdu_index = hdu_index
        with fits.open(self.filepath) as hdul:
            self.bitmask = hdul[self.hdu_index].data  # Load the data from the specified HDU

    def __len__(self):
        """
        Return the total number of entries in the dataset.
        """
        return len(self.bitmask)

    def __getitem__(self, idx):
        """
        Retrieve an item by its index.
        """
        return self.bitmask[idx]

# Path to your FITS file
filepath = '../../../projects/k-pop/spectra/apogee/dr17/aspcapStar-dr17-2M16185365+4113283.fits'

# Create the dataset
dataset = APOGEEBitMaskDataset(filepath)

# For demonstration, let's plot the whole bitmask array
plt.figure(figsize=(10, 4))
plt.plot(dataset[:], drawstyle='steps-mid')
plt.title('Bitmask Values')
plt.xlabel('Pixel Index')
plt.ylabel('Bitmask Value')
plt.show()
```

```python
import torch
from torch.utils.data import Dataset
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

class APOGEEBitMaskDataset(Dataset):
    def __init__(self, filepath, hdu_index=4):
        """
        Initialize the dataset with the path to the FITS file and the index of the HDU containing the bitmask.
        """
        self.filepath = filepath
        self.hdu_index = hdu_index
        with fits.open(self.filepath) as hdul:
            self.bitmask = hdul[self.hdu_index].data  # Ensure the data is read as integers

    def __len__(self):
        """
        Return the total number of entries in the dataset.
        """
        return len(self.bitmask)

    def __getitem__(self, idx):
        """
        Retrieve an item by its index.
        """
        return self.bitmask[idx]

# Path to your FITS file
filepath = '../../../projects/k-pop/spectra/apogee/dr17/aspcapStar-dr17-2M16185365+4113283.fits'

# Create the dataset
dataset = APOGEEBitMaskDataset(filepath)

# For demonstration, let's plot the whole bitmask array
plt.figure(figsize=(10, 4))
plt.plot(dataset[:], drawstyle='steps-mid')
plt.title('Bitmask Values')
plt.xlabel('Pixel Index')
plt.ylabel('Bitmask Value')
plt.yscale('log')  # Using logarithmic scale to better visualize bitmask values
plt.show()


print(dataset[:][2000])
```
